```python
#datacleaning 
#Drop identifier colums(they don't add analytical value)
churn_clean=churn.drop(columns=['RowNumber','CustomerId','Surname'])
#check for Duplicates
print('Duplicates:',churn_clean.duplicated().sum())
#Ensure no missing values remain
```

```python
churn_clean.info()
```


          DATA EXPLORATIONS

```python
#univariate Analysis (Histograms)
numeric_cols =['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']

for col in numeric_cols:
    plt.Figure(figsize=(6,3))
    plt.hist(churn_clean[col], bins=20, color='skyblue',edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('frequency')
    plt.tight_layout()
    plt.show()

    

,
```

```python
# catigorical Feature counts
cat_cols =['Geography','Gender','HasCrCard','Exited']
for col in cat_cols:
    plt.Figure(figsize=(5,3))
    counts =churn_clean[col].value_counts()
    plt.bar(counts.index.astype(str),counts.values,color='lightgreen',edgecolor='black')
    plt.title(f'Count of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

```

```python
# Bivarita Analysis (churn vs Gender & Geography)
churn_by_gender =churn_clean.groupby('Gender')['Exited'].mean()
plt.Figure(figsize=(4,3))
plt.bar(churn_by_gender.index,churn_by_gender.values,color='orange',edgecolor='black')
plt.title('Average churn Rate by churn_by_Gender')
plt.ylabel('churn Rate')
plt.tight_layout()
plt.show()
```

```python
Churn_by_geo =churn_clean.groupby('Geography')['Exited'].mean()
plt.figure(figsize=(5,3))
plt.bar(Churn_by_geo.index,Churn_by_geo.values,color='purple',edgecolor='black')
plt.title('Average churn Rate by Geography')
plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()
```
